```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import joblib
```

# Load Data Set

```python
dt=pd.read_csv("/kaggle/input/datasets/naveedarslan/telcc-coustmer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df=pd.DataFrame(dt)
print(df)
```

# PREPROCESSING

```python
new_df=df.drop(columns=["customerID"])
print(new_df)
```

```python
#SPLIT DATA
features=new_df.drop(columns=['Churn'])
target=new_df['Churn']

#SPLIT INTO TRAIN TEST
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
#SPLIT THE DATA INTO CATEGORICAL AND NUMERIAL 
cat=x_train.select_dtypes(include='object').columns
```

# Apply column Transformer

```python
call_trans=ColumnTransformer([
    ("Cat",OneHotEncoder(handle_unknown='ignore'),cat)
])
x_train_encoded = call_trans.fit_transform(x_train)
X_test_encoded = call_trans.transform(x_test)
```

# Creat pipline And Do Hyperparameter Tunning

```python
#CREAT PIPELINE
pip=Pipeline([
    ("pre",call_trans),
    ('model',LogisticRegression() ),  
])
```

# GRIDSEARCH

```python
#GRIDSEARCH
param_grid = {
    'model__C': [0.01, 0.1, 1, 10] ,       # Regularization strength
    'model__penalty': ['l1', 'l2'],        # Type of regularization
    'model__solver': ['liblinear']         # Required for l1
}
```

# GRIDSEARCHCV

```python
# APPLY GRIDSEARCHCV
grid = GridSearchCV(pip, param_grid, cv=5,scoring='accuracy')

# Fit model
grid.fit(x_train, y_train)
prediction=grid.predict(x_test)
print(prediction)
```

# EVALUTION

```python
ad=accuracy_score(prediction,y_test)
cr=confusion_matrix(prediction,y_test)
cl_r=classification_report(prediction,y_test)
print(ad)
print(cr)
print(cl_r)
```

# EXPORT THE MODEL

```python
joblib.dump(grid.best_estimator_, "churn_pipeline.pkl")
```
